<h2> #9 Membagi Data Frame menjadi dua secara acak

```python
# Import Modules
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

<h2> Persiapan Data Frame

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                  columns=cols)
df
```

<h2> Membagi Data Frame menjadi dua secara acak berdasarkan proporsi tertentu

```python
df.shape
```

```python
proporsi = 0.7
df_1 = df.sample(frac=proporsi)
df_2 = df.drop(df_1.index)

print(f'df_1 shape: {df_1.shape}')
print(f'df_2 shape: {df_1.shape}')
```

<h1> #10 Mengganti nama (label) kolom pada Data Frame berdasarkan pola

```python
# Import Modules
import pandas as pd

print(pd.__version__)
```

```python
# Persiapan Data Frame

df = pd.read_csv('F:/Jessica/Downloads/titanicfull.csv') 
df.columns = ['Pclass', 'Survival status', 'full name','Sex',' age',
              'Sib SP', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked']
df_backup = df.copy(deep=True)

df.head()
```

<h3> Menggunakan lowercase untuk nama kolom dan mengganti spasi dengan _

```python
df.columns = df.columns.str.replace(' ','_').str.lower()
df.head()
```

<h3> Memangkas kelebihan spasi pada nama kolom

```python
df = df_backup.copy(deep=True)

df.columns = df.columns.str.lower().str.strip().str.replace(' ','_')
df.head()
```

<h1>#11 Melakukan seleksi kolom & baris pada data frame menggunakan loc

```python
# Import Modules

import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
# Persiapan data frame

n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1,20, size=(n_rows,n_cols)),
                  columns=cols)
df
```

<h3> Seleksi kolom dan baris menggunakan loc

```python
df.loc[[0,3,4],['B','E']]
```

<h3> Seleksi baris dengan kondisi

```python
df.loc[df['B']>10, ['B','D','E']]
```

<h3> Slicing data frame dengan loc

```python
df.loc[0:4, 'B':'D']
```

<h1>#12 Membentuk kolom bertipe datetime dari sejumlah kolom lain pada data frame

```python
# Import Modules

import pandas as pd

print(pd.__version__)
```

```python
# Persiapan data frame

data = {'day':[1,2,10,25,12],
        'month':[1,2,4,5,6],
        'year':[2000,2001,2010,2015,2020]}

df = pd.DataFrame(data)
df
```

<h3> Membentuk kolom bertipe datetime

```python
df['penanggalan'] = pd.to_datetime(df[['day','month','year']])
df
```
